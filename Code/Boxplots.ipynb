## Generating box plots to display results for paper

```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
box_data = pd.DataFrame(np.array([[0.838,0.847,0.855,0.86], [0.837,0.837,0.856,0.849], [0.866,0.841,0.846,0.847], [0.806,0.835,0.856,0.859], [0.838,0.836,0.854,0.849]]),
                   columns=['Male', 'MaleNoProfanity', 'Female','FemaleNoProfanity'])
plot = sns.boxplot(data = box_data,width=0.5,fliersize=5)
plot.set_title('Male Test Data')
plot.set_ylabel('Sensitivity')
plot.set_xlabel('Model')
plt.show()
```
